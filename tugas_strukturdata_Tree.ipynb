```python
# Latihan/Tugas 1 Program Struktur Binary Tree Manual
# Struktur Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root_data):
        self.root = Node(root_data)

    def insert_left(self, current_node, data):
        if current_node.left is None:
            current_node.left = Node(data)
        else:
            new_node = Node(data)
            new_node.left = current_node.left
            current_node.left = new_node

    def insert_right(self, current_node, data):
        if current_node.right is None:
            current_node.right = Node(data)
        else:
            new_node = Node(data)
            new_node.right = current_node.right
            current_node.right = new_node

    def pre_order_traversal(self, node):
        if node is not None:
            print(node.data)
            self.pre_order_traversal(node.left)
            self.pre_order_traversal(node.right)

# Mengambil input nama dan NPM
nama = "nadhifatus zalfa az zanirah"  # Nama Anda
npm = "2410506012"  # NPM Anda

# Menghitung jumlah huruf dan vokal
jumlah_huruf = len(nama.replace(" ", ""))  # Menghitung huruf tanpa spasi
jumlah_vokal = sum(1 for char in nama.lower() if char in 'aiueo')

# Mengambil tanggal lahir (DD-MM)
tanggal_lahir = "12-01"  # Tanggal lahir Anda
tanggal, bulan = tanggal_lahir.split('-')

# Membangun binary tree
tree = BinaryTree(jumlah_huruf)  # Node root adalah jumlah huruf
tree.insert_left(tree.root, jumlah_vokal)  # Anak kiri: jumlah huruf vokal
tree.insert_right(tree.root, int(npm[-1]))  # Anak kanan: digit terakhir dari NPM
tree.insert_left(tree.root.left, int(tanggal))  # Anak kiri dari node kiri: tanggal
tree.insert_right(tree.root.left, int(bulan))  # Anak kanan dari node kiri: bulan

# Menampilkan hasil pre-order traversal
print("Struktur tree ditampilkan menggunakan pre-order traversal:")
tree.pre_order_traversal(tree.root)
```

```python
# Tugas 2 Program Binary Search Tree(BST)

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursively(self.root, data)

    def _insert_recursively(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursively(node.left, data)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursively(node.right, data)

    def in_order_traversal(self, node):
        if node is not None:
            self.in_order_traversal(node.left)
            print(node.data)
            self.in_order_traversal(node.right)

    def pre_order_traversal(self, node):
        if node is not None:
            print(node.data)
            self.pre_order_traversal(node.left)
            self.pre_order_traversal(node.right)

    def post_order_traversal(self, node):
        if node is not None:
            self.post_order_traversal(node.left)
            self.post_order_traversal(node.right)
            print(node.data)

    def search(self, data):
        return self._search_recursively(self.root, data)

    def _search_recursively(self, node, data):
        if node is None or node.data == data:
            return node
        if data < node.data:
            return self._search_recursively(node.left, data)
        return self._search_recursively(node.right)

# Data yang akan dimasukkan ke dalam BST
npm = "012"  # Tiga digit terakhir NPM
tanggal_lahir = "12-01"  # Tanggal lahir
ascii_values = [ord('N'), ord('A')]  # ASCII dari huruf 'N' dan 'A'
angka_favorit = 12  # Angka favorit

# Menggabungkan semua data
data_list = [int(npm), int(tanggal_lahir.split('-')[0]), int(tanggal_lahir.split('-')[1])] + ascii_values + [angka_favorit]

# Membangun BST
bst = BST()
for data in data_list:
    bst.insert(data)

# Menampilkan hasil traversal
print("Hasil Inorder Traversal:")
bst.in_order_traversal(bst.root)

print("\nHasil Preorder Traversal:")
bst.pre_order_traversal(bst.root)

print("\nHasil Postorder Traversal:")
bst.post_order_traversal(bst.root)

# Mencari elemen
search_value = 12  # Misalnya kita mencari angka favorit
found_node = bst.search(search_value)

if found_node:
    print(f"\nAngka {search_value} ditemukan dalam BST.")
else:
    print(f"\nAngka {search_value} tidak ditemukan dalam BST.")
```
